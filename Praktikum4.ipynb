```python
# mengimport dan menampilkan dataset

import pandas as pd
mydata = pd.read_csv("GroceryStoreDataSet.csv", names = ['products'], sep = ',')
mydata.head()
```

```python
mydata.values
```

```python
data = list(mydata["products"].apply(lambda x:x.split(",") ))
data
```

```python
!pip install mlxtend
```

```python
# melakukan proses dataset dan menyimpan nya dalam dataFrame 
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encode = te.fit(data).transform(data)

mydata = pd.DataFrame(data_encode, columns=te.columns_)
mydata = mydata.astype(int)
mydata
```

```python
# menjalankan fungsi rules
mydata = mydata.astype(int)
mydata = mydata.astype(bool)
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(mydata, min_support=0.2, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```
