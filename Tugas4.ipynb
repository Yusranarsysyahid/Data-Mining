```python
# 1Mengimpor library yang diperlukan
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Menangani nilai yang hilang jika ada
mydata = mydata.dropna(subset=["Name"])

# Mengubah kolom 'Name' menjadi list dari list
data = mydata['Name'].apply(lambda x: [x])

# Menggunakan TransactionEncoder untuk mengonversi data menjadi format biner
te = TransactionEncoder()
data_encoded = te.fit(data).transform(data)
mydata_encoded = pd.DataFrame(data_encoded, columns=te.columns_).astype(bool)
frequent_itemsets = apriori(mydata_encoded, min_support=0.01, use_colnames=True)  # Menurunkan min_support

# Menghitung aturan asosiasi
if not frequent_itemsets.empty:
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
    print("Aturan Asosiasi yang Ditemukan:")
    print(rules)
else:
    print("Tidak ada itemset yang ditemukan dengan min_support yang diberikan.")
```
